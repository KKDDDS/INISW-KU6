```python
rating2 = rating[['user id', 'movie id', 'rating']]
```

```python
rating2.pivot(index='user id', columns='movie id', values='rating')
```

```python
rating3 = rating2.set_index(['user id','movie id']).unstack().fillna(0)
```

```python
rating4 = rating3.T
```

```python
rating5 = rating4.corr()
```

```python
rating5[3].sort_values(ascending=False)
```

```python
rating5[3].nsmallest(5)
```

```python
rating3.to_numpy()
```

```python
def sim_me(user_id, n=5):
    return rating5[user_id].nlargest(n).index
```

```python
sim_me(423)
```
