```python
#numpy: 숫자를 다루기 위한 숫자 컨셉 - homogeneous 컨셉, C지원 -> vectorized 지원
```

```python
def test_sum(a,b):
    return a+b
```

```python
test_sum([1,2,3],[4,5,6])
```

```python
import numpy as np #수학적 컨셉 지원
```

```python
@np.vectorize #decorator
def test_sum(a,b):
    return a+b
```

```python
test_sum([1,2,3],[4,5,6])
```

```python
#vector: 1차 matrix: 2차, tensor: 3차 이상
```
